```python
from facebook_scraper import get_posts
import pandas as pd
import numpy as np
import time
```

```python
FANPAGE_LINK ="internetexplorerbeta"
FOLDER_PATH = "Data"
COOKIE_PATH = "tet.txt"

PAGES_NUMBER = 500 # Number of pages to crawl
```

```python
post_list = []
count = 0
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
# L∆∞u tr·ªØ c√°c t√™n c·ªôt b·∫°n mu·ªën gi·ªØ l·∫°i
desired_columns = ["comments", "reactions", "timestamp", "post_id", "post_text", "user_id", "time", "shares", "post_url",
                              "image", "images_description", "likes", "username", "user_url", "is_live",
                              "reaction_count", "page_id"]
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "timestap": True,"post_id": True, "post_text":True,"user_id": True, "time":True,"shares":True,"post_url":True,
                              "image":True, "images_description":True, "likes":True, "username":True, "user_url":True,"reaction_count":True," page_id":True, "header":headers},
                    pages=PAGES_NUMBER, cookies=COOKIE_PATH):
    # L·ªçc c√°c c·ªôt d·ª±a tr√™n t√™n c·ªôt mong mu·ªën
    filtered_post = {key: value for key, value in post.items() if key in desired_columns}
    post_list.append(filtered_post)
    print(filtered_post)
    time.sleep(10)
    count += 1
    if count == 1000:
        break
```

```python
print(len(post_list))
```

```python

# Initialize dataframe to scrape Facebook post
post_df_full = pd.DataFrame(columns=post_list[0].keys(), index=range(len(post_list)), data=post_list)

# To df
path=FOLDER_PATH + FANPAGE_LINK + ".csv"
post_df_full.to_csv(path, index=False)
print(path)
```
